# Problema 6

Utilizará las siguientes funciones:


Donde t es el número de años completos de antigüedad de los trabajadores, y las funciones determinan el valor del bono a pagar en miles de pesos según el cargo.

* Si los Administrativos recibieron un bono de 95 mil pesos, ¿Qué monto recibieron los Cargos superiores para esa misma cantidad de años de antigüedad?
* Interprete la pendiente del modelo lineal


```python
import sympy as sp
#Bono administrativo
def A(x):
    A = 4*x + 3*(x**2)
    return A 

#Bono Cargos Superiores
def S(x):
    S = 16 + 26*x
    return S


x = sp.Symbol("x")
ecuacion = A(x)-S(x)
solucion = sp.solve(ecuacion,x)
print(solucion)

print(f"Tienen que haber {solucion[1]} años de antiguedad para que el monto del bono sea el mismo.")


#Ver los años que pasaron para que los Administrativos hayan recibido 95mil pesos
#Bono administrativo
def A(x):
    A = 4*x + 3*(x**2)
    return A 

x1 = sp.Symbol("x1")
ecuacion1 = A(x1)-95
solucion1 = sp.solve(ecuacion1, x1)
print(f"Para que el Bono administrativo sea de 95mil tienen que pasar {solucion1[1]} años.")

#Ahora sacar el monto que recibieron los Cargos Superiores con los años
#Bono Cargos Superiores
def S(x):
    S = 16 + 26*x
    return S

print(f"El bono que recibieron los cargos superiores en 5 años fue de {S(5)}mil.")

#Bono Administrativo: por cada año aumentará el bono en 4mil.
#Bono Cargos Superiores: por cada años aumentará el bono en 26mil


```

# Problema 8

La utilidad de una empresa, en miles de dólares, transcurridos t meses de funcionamiento está dada por la función 
* ¿Después de cuántos años la utilidad de la empresa será de 75,85 miles de dólares


```python
import sympy as sp
import numpy as np

e = np.e

def G(x):
    G = 120 - (120*e)**-0.02*x
    return G


x = sp.Symbol("x")
ecuacion = G(x) - 75.85
solucion = sp.solve(ecuacion,x)
print(solucion)
print(round(solucion[0]))
print(f"la cantidad de años para que la utilidad de la empresa sea de 75.85 millones es de 4 años ")
```

# Problema 9

Se desea pedir un crédito de $20.000.000 en dos instituciones financieras, la tasa de interés de cada una de ellas está dada por las siguientes funciones:


Donde t corresponde al número de años al cuál se pedirá el crédito y las funciones determinan la tasa de interés en porcentaje.
* Grafique en solo gráfico ambas funciones, indique nombre de los ejes (unidades de medida), título del gráfico y una Leyenda para cada función.
* Determine, ¿a cuántos años debe pedir el préstamo para que la tasa de interés sea la misma en ambas financieras?
* ¿Cuándo le conviene pedir el préstamo en la Financiera N°2?


```python

import matplotlib.pyplot as plt
import numpy as np


def I(t):
    I= -0.2*t+3
    return I
def L(t):
    L=-2/15*t+8/3
    return L




# crear gráfico
t1= np.arange(0, 10, 0.1)
t2= np.arange(0, 10, 0.1)
plt.plot(t1, I(t1), label="Financiera 1 I(t)")
plt.plot(t2, L(t2), label="Financiera 2 L(t)")


plt.ylim(0,4)
plt.title("taza de interes en porcentaje en funcion del tiempo en años")
plt.xlabel("tiempo en años") #nombre de valores
plt.ylabel("% de interes")


#modificar ejes:
ax=plt.gca()
ax.spines[["left", "bottom"]].set_position(("data", 0))
ax.spines[["right", "top"]].set_visible((False))




plt.grid() #recuadros
plt.legend() #Nombre de la linea (eje)
plt.show() #Llamada al gráfico

print("Se debe pedir préstamos a los 5 años para quela tasa de interés sea la misma en ambas financieras.")
print("Le conviene pedir un prestamo en la Financiera 2 antes de los 5 años.")
```
