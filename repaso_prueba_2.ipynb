```python
import numpy as np
import sympy as sp

def b(n):
    b = (2*(n**3))-(12*(n**2))
    return b

x = sp.Symbol("x")
ecuacion = b(x)-13284800
solucion = sp.solve(ecuacion,x)
print(f"El valor de n para que sea {b(190)} tiene que ser {solucion[0]}")

def c(n):
    c = 0.5*(n**3)-0.2*n
    return c

x1 = sp.Symbol("x1")
ecuacion1 = c(x1)-58814.7
solucion1= sp.solve(ecuacion1,x1)
print(f"El valor de n para que sea {c(49)} tiene que ser {solucion1[0]}")
```

```python
import numpy as np
#Matriz M - 2x3 - plantas (1 y 2) x dispositivos (tabletas, laptops, smartphones)
M = np.array([[500, 300, 400],[600, 200, 350]])

#Matriz N - 3x2 - dispositivos (tablets, laptops, smartphones) x años (2023, 2024)
N = np.array([[120, 110],[200,190],[150,140]])

#a)
#  a)M+N = 2x3 + 3x2 = no se puede sumar
#  b)NxM = 3x2 * 2x3 = me dará una matriz 2x2 (planta x año) 
#  c)MxN = 2x3 * 3x2 = me dará una matriz 3x3 (dispositivos x dispositivos), no tiene sentido.
#  d) M.T X N = 3x2 * 3x2 = no, porque filas de M.T no es igual a las columnas de N

print("a) La operación que me permite obtener el costo total de producción por planta en cada año es 'c) M x N', porque me dará una matriz planta x año, lo cual representa la demanda total " \
"de la planta por año.")
P = M@N

p22 = P[2-1][2-1]
print(f"\nb) p22 = {p22}. Lo cual me dice que en la planta 2 hubo un costo total de producción de ${p22} en el año 2024")
```

```python
import numpy as np
#Matriz A - 2x3 = gasto de producción (costo mat., mano de obra) x clases de producto (A, B, C)
#Estimado del costo a producir un solo artículo de cada clase (en dólares)
A = np.array([[0.20, 0.40, 0.25],[0.40, 0.50, 0.35]])


#Matriz B - 3x4 = clases de producto (A, B, C) x estaciones de año (verano, otoño, invierno, primavera)
#Estimado de la cantidad de artículos de clada clase que se podrían producir en cada estación del año.
B = np.array([[5000, 5500, 5500, 5000],[3000, 3600, 3400, 3200],[6800, 7200, 7000, 7000]])

T = A@B
print(f"a) Acá se muetra la matriz T - costos de producción x estaciones del año: \n{T}")

t12 = T[1-1][2-1]
print(f"\nb) t12 = {t12}. Me dice que en los costos de materiales durante otoño son de ${t12}.")

t23 = T[2-1][3-1]

print(f"\nc) El costo total asociado a la mano de obra en invierno estaría representado por el elemento t23, lo cual equivale a ${t23}")
```

```python
import numpy as np

# x: minutos que software A estára operando
# y: minutos que software B estará operando

A = np.array([[2, 3], [1, 4]])
B = np.array([[52],[56]])

A_inv = np.linalg.inv(A)

X = A_inv@B

#Comprobar si está bien
B_comprobar = A@X
print(f"Comprobar si está bien (tienen que ser iguales): \nB_comprobar: \n{B_comprobar} \nB: \n{B}")
print("Todo ok.")

print(f"El Software A tiene que estar operando {X[0]} minutos. Y el Software B tiene que estar operando {X[1]} minutos para cumplir con los requerimientos de procesamiento de solicitudes.")

```
