# Problema 1

La fábrica de bicicletas Expend produce dos modelos de bicicletas, $M$ y $N$. La matriz $A$ nos indica las cantidades de acero y aluminio que requiere cada bicicleta, expresadas en kg según el modelo a producir, mientras que la matriz $B$ nos entrega un reporte de la cantidad de bicicletas producidas de ambos modelos para los dos primeros meses del año pasado.

<br>

\begin{equation*}
\begin{matrix}
& \begin{matrix}
 & \ \ \ \  M & N
\end{matrix}  \\
&A= \begin{pmatrix}
1 & 2 \\
3 & 2 \\
\end{pmatrix} & \ \begin{matrix}
Acero\\
Aluminio
\end{matrix}\\
\end{matrix}
\ \ \ \ \ \ \ \ \ \ \ \ \ \
\begin{matrix}
& \begin{matrix}
& \ \ \ \ \ \ \ ene & feb &
\end{matrix} & \\
&B= \begin{pmatrix}
290 &  312 \\
 345 &  413 \\
\end{pmatrix} & \begin{matrix}
M\\
N
\end{matrix}\\
\end{matrix}
\end{equation*}

<br>

1. Indique el orden de cada matriz. ¿Es posible determinar la multiplicación entre ambas matrices? Justifique.
2. Determine utilizando *Python* la matriz $C = A \cdot B$
3. Indique e interprete el valor de los elementos $c_{21}$ y $c_{11}$.

```python
import numpy as np

A = np.array([[1,2],[3,2]])
B = np.array([[290,312],[345,413]])

#1
#El orden de A es 2x2
#El orden de B es 2x2
#El orden de la matiz C es 2x2 (filas de primera matriz por columnas de segunda matriz)

print("Si es factible multiplicar, ya que la cantidad de columnas de la primera matriz es igual a la cantidad de filas de la segunda matriz.")

C = A@B
print(f"2.\n{C}")

c21 = C[2-1][1-1]
print("c21 = ")
c11 = C[1-1][1-1]

print(f"El elemento c21 es {c21}. Se utilizaron {c21} kg de aluminio en el mes de enero para la fabricación de ambos modelos.")
print(f"El elemento c11 es {c11}. Se utilizaron {c11} kg de aluminio en el mes de enero para la fabricación de ambos modelos.")
```

\begin{split}
V &= \begin{pmatrix}
 1300 &  900  &  800  \\

\\
P &= \begin{pmatrix}
 4200 &  1500 & 3200 \\

<br>

\begin{split}
h + m & = 60 \\
0,16h + 0,2m & = 11
\end{split}

<br>

$\hspace{1cm}$ Se puede escribir matricialmente como:

<br>

\begin{split}
\begin{pmatrix}
 1  &  1 \\
 0,16 & 0,2 \\
\end{pmatrix} \cdot \begin{pmatrix}
 h \\
 m \\
\end{pmatrix} = \begin{pmatrix}
 60 \\
 11 \\
\end{pmatrix}
\end{split}

<br>

* Utiliza la matriz inversa para encontrar los valores de $h$ y $m$ que permiten satisfacer el sistema de ecuaciones.

<br>

2m + 4p & = 2900 \\

<br>
**b)**
\begin{split}
n + v & = 160 \\

<br>

1. Escribe ambos sistemas de ecuaciones de forma matricial.

# Problema 12

En una granja hay $132$ animales entre patos y vacas. En total hay $402$ patas. El sistema que modela la situación planteada es:

<br>

\begin{split}
p + v & = 132 \\
2p + 4v & = 402
\end{split}

<br>

1. ¿Qué representan las incógnitas del sistema?
2. ¿Cuántos patos hay?

```python
#1
# la "p" representa patos
# la "v" representa vacas
# 132 representa el total de animales

#2p + 4v = 402 --> cada pato tiene 2 patas y las vacas tienen 4 patas = entre los dos animales hay un total de 402 patas

#2
import numpy as np
A = np.array([[1,1],[2,4]])
B = np.array([[132],[402]])

A_inv = np.linalg.inv(A)

X = A_inv@B

#Comprobar
B_comprobar = A@X
print(f"Comprobar:\n{B} \ntiene que ser igual a \n{B_comprobar}")

print(f"\n2.\nHay un total de {X[0,0]} patos y un total de {X[1,0]} vacas.")
```

El sistema que modela el problema es:

<br>

1. ¿Qué representa cada incógnita en el contexto del problema?
